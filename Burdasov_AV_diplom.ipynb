### 2.3 Очистка и преобразование данных ###

Загружаем необходимые библиотеки и файл данных:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('uln_08-23.csv', delimiter=';')
df.head()
```

Посмотрим информацию о нашем загруженном датасете

```python
df.info()
```

```python
duplicateRows = df[df.duplicated ()]
print(duplicateRows)
```

Как видно, набор данных, который был мною получен достаточно полный, пропущенных практичеки значений не имеется - только по колонкам 'Персона для контактов', 'Ссылки на картинки','Контактное лицо' (автор объявления)  но они нам будут не нужны.
Дубликатов данных так же не имеется. 
По колонке  'Метро/Район' работу проведем позднее.

Удалим все персональные данные, данные, относящиеся к сайта и ненужные данные типа 'Расстояние до метро, км' (все равно в Ульяновске нет метро). 
Колонку 'Метро/Район'переименуем в 'Район':

```python
df=df.drop(['Телефон', 'Оператор', 'Контактное лицо (автор объявления)', 'Тип автора',
            'Тип объявления', 'Категория1', 'Категория2', 'ID на сайте',	'Источник',
            'Персона для контактов', 'URL', 'Ссылки на картинки', 'Регион мобильного телефона',
            'Номер подменён', 'Расстояние до метро, км'], axis=1)
df.rename(columns = {'Метро/Район':'Район'}, inplace = True )
df.head(0)
```


Посмотрим, какие типы квартир имеются в нашем наборе данных:

```python
df['first'] = df['Название'].str.split(' ').str[0]
df.groupby('first').count().iloc [:, 0]
```

```python
df.loc[~df['Название'].astype(str).str.contains(r'^(?:[1-9])')]
```


Удалим объяевления о продаже долей в квартире и аукционы:

```python
df=df[df['Название'].str.contains("Доля|Аукцион")== False]
```


Теперь займемся пропущенными данныим по колонке 'Район'. У нас имеется 45 таких объявлений. Посмотрим на них:

```python
df[df['Район'].isnull()]
```

Сразу же видно, что в набор данных пополи объявления не по городу Ульяновску а по различным населенным пунктам Ульяновской области. Удалим все объявления, где адрес содержит 'Ульяновская область,':

```python
df=df[df['Адрес'].str.contains('Ульяновская область,')== False]
```

Выведем оставшиеся объявления с пустым значением'Район':

```python
df[df['Район'].isnull()]
```

У нас осталось 10 таких объявлений, их можно просмотреть и поправить "вручную". У трех объявлений есть адрей и по названию улицы можно определить и внести название района.

```python
df.at [2333, 'Район'] = 'Заволжский'
df.at [4215, 'Район'] = 'Засвияжский'
df.at [3238, 'Район'] = 'Ленинский'

df[df['Район'].isnull()]
```

Так как у всех объявлений есть географические координаты - у оставшихся проверим координаты через какой-нибудь геосервис. Я воспользовался https://geotree.ru. 
Все объявления оказались тоже за пределами города Ульяновск, поэтому удаляем их из нашего датасета.

```python
df=df.dropna()
df[df['Район'].isnull()]
```

Этап очистки окончен и теперь переходим к преобразованию данных. 
В объявлениях все данные квартиры записаны в одном текстовом поле. С помощью сплит-оператора выделим из этой строки данные о площади, этаже  и количеству комнат в отдельные колонки.

```python
new_df = df['Название'].str.split(',',expand=True)
new_df.columns=['квартира','площадь','этаж/высота дома']
new_df['этаж'] = new_df['этаж/высота дома'].str.split('/').str[0].astype(int)
new_df['кол-во комнат'] = new_df['квартира'].str[0]
new_df
```

Так как в датасете находились объявления "Апартаменты-студия", "Квартира-студия", "Своб." - о квартирах свободной планировки (без определенного количества комнат), в колонке количества комнат появились значения "А", "К", "С". 
Заменим эти буквы на 0.

```python
new_df['кол-во комнат'] = new_df['кол-во комнат'].replace('С','0').replace('А','0').replace('К','0').astype(int)
new_df.sort_values(by=['кол-во комнат'])
```


Колонку с указанием площади из строкового формата переведем в числовой:

```python
new_df['площадь'] = new_df['площадь'].str.replace('м²','')
```

```python
new_df.sort_values(by=['площадь'])
```

```python
new_df['площадь'] = new_df['площадь'].astype(float)
```

Проверим результат преобразований:

```python
new_df.info()
```

Результаты наших преобразований сводим в новый датасет - df_realty, который и будет использоваться в дальнейшей работе по проведению анализа объявлений.

```python
df_realty = pd.concat([df,new_df],axis=1)
df_realty
```

```python
x = df_realty.loc[:, ('Район', 'площадь', 'кол-во комнат', 'этаж')]
y = df_realty.loc[:, 'Цена']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
x_train['Район'] = pd.factorize(x_train['Район'])[0]
x_test['Район'] = pd.factorize(x_test['Район'])[0]
```

```python
mms = MinMaxScaler()
mms.fit(x_train)
x_train = mms.transform(x_train)
x_test = mms.transform(x_test)

x_train 
```

```python
x_test
```

```python
# Создадим модель
model_lr = LogisticRegression()
model_lr.fit(x_train, y_train)# обучение
y_pred = model_lr.predict(x_test)# предсказание 
```

```python
print(f"Training set score: {model_lr.score(x_train, y_train):.3%}")
print(f"Testing set score: {model_lr.score(x_test, y_test):.3%}\n")
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print('mse: %.3f, mae: %.3f' % (mse, mae))
```
